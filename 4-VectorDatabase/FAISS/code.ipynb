```python
from langchain_huggingface import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
embeddings.embed_query("Hello AI")
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sentences = [
    
    "The children are playing football in the field.",
    "He is learning Python for data science.",
    "The weather today is sunny and bright."
]

query = "Someone is writing code in Python."
```

```python
sentences_embed = embeddings.embed_documents(sentences)
sentences_embed
```

```python
query_embed = embeddings.embed_query(query)
query_embed
```

```python

cosine_similarity([query_embed],sentences_embed)
```

```python
from sklearn.metrics.pairwise import euclidean_distances as l2
```

```python
l2([query_embed],sentences_embed)
```

```python
from langchain_core.documents import Document

document_1 = Document(
    page_content="Just brewed some fresh coffee and getting ready to start coding for the day.",
    metadata={"source": "tweet"},
)

document_2 = Document(
    page_content="A major earthquake hit Japan early this morning, causing widespread damage and power outages.",
    metadata={"source": "news"},
)

document_3 = Document(
    page_content="Exploring how vector databases like FAISS can speed up AI search systems.",
    metadata={"source": "website"},
)

document_4 = Document(
    page_content="Feeling blessed to have friends who support me no matter what!",
    metadata={"source": "tweet"},
)

document_5 = Document(
    page_content="The new electric vehicle from Tesla promises a 600-mile range on a single charge.",
    metadata={"source": "news"},
)

document_6 = Document(
    page_content="Python remains one of the top programming languages in 2025 for AI and data science.",
    metadata={"source": "website"},
)

document_7 = Document(
    page_content="LangChain now integrates with Mistral and Llama 3 models for better LLM performance.",
    metadata={"source": "news"},
)

document_8 = Document(
    page_content="Just launched my first AI app using Streamlit and LangChain! Feels amazing!",
    metadata={"source": "tweet"},
)

document_9 = Document(
    page_content="The football world cup final was a thrilling match that kept fans on edge until the last minute.",
    metadata={"source": "news"},
)

document_10 = Document(
    page_content="Working late nights isn't fun, but finishing a tough project sure feels rewarding.",
    metadata={"source": "tweet"},
)

documents = [
    document_1,
    document_2,
    document_3,
    document_4,
    document_5,
    document_6,
    document_7,
    document_8,
    document_9,
    document_10,
]
```

```python
index=faiss.IndexFlatIP(384)
vector_store=FAISS(
    embedding_function=embeddings,
    index=index,
    docstore=InMemoryDocstore(),
    index_to_docstore_id={},
)
```

```python
vector_store.add_documents(documents=documents)
```

```python
vector_store.similarity_search(
    "LangChain provides abstractions to make working with LLMs easy",
    k=2 #hyperparameter
    
)
```

```python
# creating the retriever using this vector store 
retriever=vector_store.as_retriever(search_kwargs={"k": 3})
```

```python
retriever.invoke("LangChain provides abstractions to make working with LLMs easy")
```

```python
#saving vector store in my local 
vector_store.save_local("faiss index")
```
