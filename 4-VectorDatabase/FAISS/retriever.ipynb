```python
from langchain_community.document_loaders import PyPDFLoader
```

```python
FILE_PATH = r"E:\Agentic-AI-code\4-VectorDatabase\llama2.pdf"
loader=PyPDFLoader(FILE_PATH)
pages = loader.load()
```

```python
len(loader.load())
```

```python
pages = []
async for page in loader.alazy_load():
    pages.append(page)
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,#hyperparameter
    chunk_overlap=50 #hyperparemeter
)
```

```python
split_docs = splitter.split_documents(pages)
```

```python
len(split_docs)
```

```python
import faiss
from langchain_community.docstore.in_memory import InMemoryDocstore
from langchain_community.vectorstores import FAISS
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
embeddings=HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
index=faiss.IndexFlatIP(384)
vector_store=FAISS(
    embedding_function=embeddings,
    index=index,
    docstore=InMemoryDocstore(),
    index_to_docstore_id={},
)
```

```python
vector_store.add_documents(documents=split_docs)
```

```python
retriever=vector_store.as_retriever(
    search_kwargs={"k": 6} #hyperparameter
)
```

```python
retriever.invoke("what is llama model?")
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
model=ChatGoogleGenerativeAI(model='gemini-2.5-pro' ,  api_key=api_key)
model
```

```python
import os
LANGSMITH_API_KEY=os.getenv("LANGSMITH_API_KEY")
```

```python

from langsmith import Client
client = Client(api_key=LANGSMITH_API_KEY)
prompt = client.pull_prompt("rlm/rag-prompt", include_model=True)
```

```python

print(prompt)
```

```python
import pprint
pprint.pprint(prompt.messages)
```

```python
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough
```

```python
def format_docs(docs):
    return "\n\n".join(doc.page_content for doc in docs)
    
```

```python
rag_chain = (
    {"context": retriever | format_docs, "question": RunnablePassthrough()}
    | prompt
    | model
    | StrOutputParser()
)
```

```python
rag_chain.invoke("what is llama model?")
```
